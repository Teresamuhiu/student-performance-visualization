# Student Performance Visualization Project

This notebook explores the Students Performance dataset using Python's Pandas, Seaborn, and Matplotlib libraries. The goal is to identify score patterns, gender differences, and relationships between subjects. Each chart includes educational insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path
file_path = '/content/drive/My Drive/StudentsPerformance.csv'

# Load the CSV
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

### Visualization Charts

```python
#Histogram of Math Scores
plt.figure(figsize=(10, 5))
sns.histplot(df['math score'], kde=True, color='red')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

### Distribution of Math Scores – Histogram

This histogram shows how students’ math scores are spread across the dataset.

- Each bar represents the number of students who scored within a certain range (e.g., 60–70).
- The red line (KDE) helps visualize the overall shape of the data.

**What we notice:**
- Most students scored between 60 and 80.
- A few students scored very low (under 30), which creates a slight left skew in the distribution.
- The distribution is roughly bell-shaped, which is typical in exam score data.

**Why it matters:**
This tells instructors what a “typical” math score looks like and whether the class is skewed toward high or low performance. It helps identify whether the test was fair or too hard.


```python
#Boxplot: Math Score by Gender
plt.figure(figsize=(10, 5))
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

### Math Scores by Gender (Boxplot)

This boxplot compares the distribution of math scores between male and female students.

- The **line inside each box** shows the median math score.
- The **box height** shows the middle 50% of scores (interquartile range).
- The **"whiskers"** show the full range of typical scores.
- **Dots** below the box represent outliers (unusually low scores).

**What we notice:**
- Male students have a slightly higher median math score.
- Both groups have similar score spread, but female students have more low-score outliers.

**Why it matters:**
This helps us explore whether gender might be associated with different performance trends in math, which can inform how assignments or support are structured.



```python
#Correlation Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Exam Scores")
plt.show()
```

### Correlation Between Math, Reading, and Writing Scores – Heatmap

This heatmap shows the strength of the relationship between students' scores in math, reading, and writing.

- A correlation value of **1.0** means a perfect relationship.
- **Values closer to 1** mean the two scores tend to increase or decrease together.
- **Cool colors (blue)** show lower correlations; **warm colors (red)** show stronger ones.

**What we notice:**
- Reading and writing scores are very strongly correlated (0.95).
- Math is positively related to both reading (0.82) and writing (0.80), but slightly less so.

**Why it matters:**
Understanding these relationships can help instructors predict performance patterns and design cross-subject support materials. For example, students strong in reading are likely to also do well in writing.



### Key Insights:
- Most students score between 60 and 80 in math.
- Males have a slightly higher median math score; females show more low-score outliers.
- Reading and writing scores are highly correlated and good readers tend to be strong writers.
